# CIFAR 10 in Pytorch

```python
import torch
import torchvision 
import torchvision.transforms as transforms
from torch import nn
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import random_split
import torch.optim as optim
import time
```

```python
transform = transforms.Compose([transforms.Resize((32,32)),
                                transforms.ToTensor(), # convert the image to tensor so that it can work with torch
     transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
     ]) # Normailize all the images

transformAlexNet = transforms.Compose([transforms.Resize((70,70)),
                                transforms.ToTensor(), # convert the image to tensor so that it can work with torch
     transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
     ]) # Normailize all the images

transformjitterroattion = transforms.Compose([transforms.Resize((70,70)),  #resises the image so it can be perfect for our model.
                                      transforms.RandomHorizontalFlip(), # FLips the image w.r.t horizontal axis
                                      transforms.RandomRotation(10),     #Rotates the image to a specified angel
                                      transforms.RandomAffine(0, shear=10, scale=(0.8,1.2)), #Performs actions like zooms, change shear angles.
                                      transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2), # Set the color params
                                      transforms.ToTensor(), # comvert the image to tensor so that it can work with torch
     transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
     ]) # Normailize all the images
                                              
# https://www.kaggle.com/code/vikasbhadoria/cifar10-high-accuracy-model-build-on-pytorch reduce overfitting
# This takes in how much memory it would use
batch_size = 128

#train set
trainset = torchvision.datasets.CIFAR10(root = './data', train=True,
                                    download = True, transform=transform)
validset = torchvision.datasets.CIFAR10(root='./data', train=False, 
                                             download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data', train=False, 
                                             download=True, transform=transform)
#Loader
train_dl = torch.utils.data.DataLoader(trainset, batch_size= batch_size,
                                          shuffle= True, num_workers=2)
valid_dl = torch.utils.data.DataLoader(validset, batch_size = batch_size, shuffle=False)

test_dl = torch.utils.data.DataLoader(testset, batch_size = 10000, shuffle=False) # this will test everything for the confustion matrix

#train set
trainsetalex = torchvision.datasets.CIFAR10(root = './data', train=True,
                                    download = True, transform=transformAlexNet)
validsetalex = torchvision.datasets.CIFAR10(root='./data', train=False, 
                                             download=True, transform=transformAlexNet) 
testsetalex = torchvision.datasets.CIFAR10(root='./data', train=False, 
                                           download=True, transform=transformAlexNet)                                            
#train set jitter
trainsetjitalex = torchvision.datasets.CIFAR10(root = './data', train=True,
                                    download = True, transform=transformjitterroattion)
validsetjitalex = torchvision.datasets.CIFAR10(root='./data', train=False, 
                                             download=True, transform=transformjitterroattion)                                              
                                              
                                              
#Loader
train_dlalex = torch.utils.data.DataLoader(trainsetalex, batch_size= batch_size, shuffle= True, num_workers=2)
valid_dlalex = torch.utils.data.DataLoader(validsetalex, batch_size = batch_size, shuffle=False)
test_dlalex = torch.utils.data.DataLoader(testsetalex, batch_size = batch_size, shuffle=False)

train_dlalexjit = torch.utils.data.DataLoader(trainsetjitalex, batch_size= batch_size, shuffle= True, num_workers=2)
valid_dlalexjit = torch.utils.data.DataLoader(validsetjitalex, batch_size = batch_size, shuffle=False)                                              

#classes inside cifar 10
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

Function to Draw the Loss vs no of Epochs graph

```python
def draw_loss_epochs( loss_history,val_loss_history ):
  plt.style.use('ggplot')
  plt.plot(loss_history, label='training loss')
  plt.plot(val_loss_history, label='validation loss')
  plt.title("Shows the Loss for Training and validation set")
  plt.xlabel("Epochs")
  plt.ylabel("Loss")
  plt.legend()
  plt.show()
```

Function to Draw Accuracy over epochs on a graph


```python
def draw_acc_epochs(acc_history,val_acc_history ):
  plt.style.use('ggplot')
  plt.plot(acc_history, label='training acc')
  plt.plot(val_acc_history, label='validation acc')
  plt.title("Shows the Acc for Training and validation set")
  plt.xlabel("Epochs")
  plt.ylabel("Accuracy")
  plt.legend()
  plt.show()
```

Function to draw the accuracy

```python
def accuracy_test_percent(cnnnetwork, testloader):
    correct = 0
    total = 0
    if torch.cuda.is_available():
        device = torch.device("cuda")
    else:
        device = torch.device("cpu")
    with torch.no_grad():
        for data in testloader:
            acc_images, acc_labels = data
            acc_images = acc_images.to(device)
            acc_labels = acc_labels.to(device)
            # calculate outputs by running images through the network
            outputs = cnnnetwork(acc_images)
            # the class with the highest energy is what we choose as prediction
            _, predicted = torch.max(outputs.data, 1)
            total += acc_labels.size(0)
            correct += (predicted == acc_labels).sum().item()
    print(f'Accuracy of the network on the 10000 test images: {100 * correct // total} %')
```

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from time import gmtime, strftime
import torch.nn as nn
import torch.nn.functional as F

def convultionalnnfunc(cnnnetwork ,epochcount, learningr, mom, train_loader, valid_loader, test_loader):
  net = cnnnetwork

  ## This checks if a graphics cards is avaliable 
  if torch.cuda.is_available():
    device = torch.device("cuda")
  else:
    device = torch.device("cpu")
  net.to(device)

  criterion = nn.CrossEntropyLoss()
  # Fine tuned for the learning rate
  optimizer = optim.SGD(net.parameters(), lr=learningr, momentum=mom)

  # This is to plot the loss history
  running_loss_history = []
  val_running_loss_history = []

  # This is to plot the Accuracy history
  running_corrects_history = []
  val_running_corrects_history = []
  # We iter the batch of images to display

  for epoch in range (epochcount):
    print (strftime("%Y-%m-%d %H:%M:%S", gmtime()))
    
    running_loss = 0.0
    running_corrects = 0.0
    val_running_loss = 0.0
    val_running_corrects = 0.0

    for inputs, labels in train_loader:
      inputs = inputs.to(device) # input to device as our model is running in mentioned device.
      labels = labels.to(device)
      outputs = net(inputs) # every batch images are put as an input.

      # print ("outputs ",outputs.size(), " loss", outputs.size())
      loss = criterion(outputs, labels)

      optimizer.zero_grad() #set the additional value to 0
      loss.backward() #back propagation
      optimizer.step() #update the weights 

      _, preds = torch.max(outputs, 1) # taking the highest value of prediction.
      running_loss += loss.item() 
      running_corrects += torch.sum(preds == labels.data) # calculating te accuracy by taking the sum of all the correct predictions in a batch.
    else:
      with torch.no_grad(): # we do not need gradient for validation.
        for val_inputs, val_labels in valid_loader:
          val_inputs = val_inputs.to(device)
          val_labels = val_labels.to(device)
          val_outputs = net(val_inputs)
          val_loss = criterion(val_outputs, val_labels)
          
          _, val_preds = torch.max(val_outputs, 1)
          val_running_loss += val_loss.item()
          val_running_corrects += torch.sum(val_preds == val_labels.data)
    ## Calculate the loss and acc per epoch in the training set
    epoch_loss = running_loss/len(train_loader) # loss per epoch
    epoch_acc = running_corrects.float()/ len(train_loader) # accuracy per epoch
    running_loss_history.append(epoch_loss)
    running_corrects_history.append(epoch_acc)
    
    ## Calculate the loss and acc per epoch in the validation set
    val_epoch_loss = val_running_loss/len(valid_loader) # loss per epoch
    val_epoch_acc = val_running_corrects.float()/ len(valid_loader) # accuracy per epoch
    val_running_loss_history.append(val_epoch_loss) # appending for displaying 
    val_running_corrects_history.append(val_epoch_acc)

    if ((epoch+1)%4 == 0 or epoch==0 ):
      print('epoch :', (epoch+1),"/", epochcount)
      print('training loss: {:.4f}, acc {:.4f} '.format(epoch_loss, epoch_acc.item()))
      print('validation loss: {:.4f}, validation acc {:.4f} '.format(val_epoch_loss, val_epoch_acc.item()))

    #https://www.kaggle.com/code/vikasbhadoria/cifar10-high-accuracy-model-build-on-pytorch/notebook
  
  print('Finished Training')
  #plot training loss with validation loss to stop overfitting 
  draw_loss_epochs(running_loss_history, val_running_loss_history)
  draw_acc_epochs(running_corrects_history, val_running_corrects_history)
  accuracy_test_percent(net, test_loader)

  # PATH = './cifar_net.pth'
  # torch.save(net.state_dict(), PATH)
  # net.load_state_dict(torch.load(PATH))  

  total_correct = 0
  total_images = 0
  confusion_matrix = np.zeros([10,10], int) 

  with torch.no_grad():
    for data in test_loader:
      images, labels = data
      images = images.to(device)
      labels = labels.to(device)
      outputs = net(images)
      _, predicted = torch.max(outputs.data, 1)
      total_images += labels.size(0)
      total_correct += (predicted == labels).sum().item()

      for i, l in enumerate(labels):
        confusion_matrix[l.item(), predicted[i].item()] += 1 

  model_accuracy = total_correct / total_images * 100
  print('Model accuracy on {0} test images: {1:.2f}%'.format(total_images, model_accuracy))

  print('actual/pred'.ljust(16), end='')
  for i,c in enumerate(classes):
    print(c.ljust(10), end='')
  print()
  for i,r in enumerate(confusion_matrix):
    print(classes[i].ljust(16), end='')
    for idx, p in enumerate(r):
      print(str(p).ljust(10), end='')
    print()

    r = r/np.sum(r)
    print(''.ljust(16), end='')
    for idx, p in enumerate(r):
      print(str(p).ljust(10), end='')
    print()
  plot_confusion_matrix
#https://saptarshidatta.in/2020/10/05/PyTorch_CIFAR10.html copied the confusion matrix funciton
```

```python
# LeNET Model
class Net(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1 = nn.Conv2d(3,6,5)
    self.pool = nn.MaxPool2d(2,2)
    self.conv2 = nn.Conv2d(6,16,5)
    self.fc1 = nn.Linear(16 * 5 * 5, 120)
    self.fc2 = nn.Linear(120, 84)
    self.fc3 = nn.Linear(84, 110)
  
  def forward(self, x):
    x = self.pool(F.relu(self.conv1(x)))
    x = self.pool(F.relu(self.conv2(x)))
    x = torch.flatten(x, 1) # flatten all dimensions except batch
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.fc3(x)
    # x = F.softmax(self.fc3(x),dim=1) #softmax the final Net module 
    return x
# Running with 32 number of Epochs
# https://pytorch.org/tutorials/beginner/blitz/cifar10_tutorial.html
# Args Model, Epochs, Learning Rate, Momentum 
convultionalnnfunc(Net() , 32, 0.001, 0.9, train_dl, valid_dl, test_dl)
```

Regularisation to Prevent overfitting: 

```python
## lenet regularisation
class LeNetreg(nn.Module):
  def __init__(self):
    super(LeNetreg, self).__init__()

    self.convolutional_layer = nn.Sequential(            
      nn.Conv2d(in_channels=3, out_channels=6, kernel_size=5, stride=1),
      nn.ReLU(),
      nn.MaxPool2d(kernel_size=2, stride=2, padding=0),
      nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5, stride=1),
      nn.ReLU(),
      nn.MaxPool2d(kernel_size=2, stride=2, padding=0),
      nn.Conv2d(in_channels=16, out_channels=120, kernel_size=5, stride=1),
      nn.ReLU()
    )

    self.linear_layer = nn.Sequential(
        nn.Dropout(),
        nn.Linear(in_features=120, out_features=84),
        nn.ReLU(),
        nn.Linear(in_features=84, out_features=10),
    )

  def forward(self, x):
    x = self.convolutional_layer(x)
    x = torch.flatten(x, 1)
    x = self.linear_layer(x)
    x = F.softmax(x, dim=1)
    return x

    x = self.convolutional(x)
    x = self.avgpool(x)
    x = torch.flatten(x, 1)
    x = self.linear(x)
    return torch.softmax(x, 1)
# https://datahacker.rs/018-pytorch-popular-techniques-to-prevent-the-overfitting-in-a-neural-networks/
# https://medium.com/analytics-vidhya/understanding-regularization-with-pytorch-26a838d94058
convultionalnnfunc(LeNetreg() , 80, 0.001, 0.9, train_dl, valid_dl, test_dl)
```

# LeNet with Dropout example

```python
class NetDropout(nn.Module):
  def __init__(self, n_chans1=32, debug=False):
    self.debug = debug
    super().__init__()
    self.n_chans1 = n_chans1
    self.conv1 = nn.Conv2d(3, n_chans1, kernel_size=3, padding=1)
    self.conv1_dropout = nn.Dropout3d(p=0.3)
    self.conv2 = nn.Conv2d(n_chans1, n_chans1//2, kernel_size=3, padding=1)
    self.conv2_dropout = nn.Dropout2d(p=0.3)
    self.fc1 = nn.Linear(8*8*n_chans1//2, 32)
    self.fc2 = nn.Linear(32,10)

  def forward(self, x):
    out = F.max_pool2d(torch.tanh(self.conv1(x)), 2)
    out = self.conv1_dropout(out)
    out = F.max_pool2d(torch.tanh(self.conv2(out)), 2)
    out = self.conv2_dropout(out)
    out = out.view(-1,8*8*self.n_chans1//2)
    out = torch.tanh(self.fc1(out))
    out = self.fc2(out)
    return out

# https://www.programcreek.com/python/example/107657/torch.nn.Dropout2d
```

```python
class AlexNet(nn.Module):
  def __init__(self):
    super().__init__()
    self.features = nn.Sequential(
      nn.Conv2d(3, 64, kernel_size=11, stride=4, padding=2),
      nn.ReLU(inplace=True),
      nn.MaxPool2d(kernel_size=3, stride=2),
      nn.Conv2d(64, 192, kernel_size=5, padding=2),
      nn.ReLU(inplace=True),
      nn.MaxPool2d(kernel_size=3, stride=2),
      nn.Conv2d(192, 384, kernel_size=3, padding=1),
      nn.ReLU(inplace=True),
      nn.Conv2d(384, 256, kernel_size=3, padding=1),
      nn.ReLU(inplace=True),
      nn.Conv2d(256, 256, kernel_size=3, padding=1),
      nn.ReLU(inplace=True),
      nn.MaxPool2d(kernel_size=3, stride=2),
    )
    self.avgpool = nn.AdaptiveAvgPool2d((6, 6))
    self.classifier = nn.Sequential(
        nn.Dropout(0.5),
        nn.Linear(256 * 6 * 6, 4096),
        nn.ReLU(inplace=True),
        nn.Dropout(0.5),
        nn.Linear(4096, 4096),
        nn.ReLU(inplace=True),
        nn.Linear(4096, 10)
    )
  def forward(self, x):
    x = self.features(x)
    x = self.avgpool(x)
    x = x.view(x.size(0), 256 * 6 * 6)
    logits = self.classifier(x)
    return logits
# https://github.com/rasbt/stat453-deep-learning-ss21/blob/main/L13/code/2-alexnet-cifar10.ipynb
convultionalnnfunc(AlexNet(), 32, 0.003, 0.9, train_dlalex, valid_dlalex, test_dlalex)  #learning rate 0.003
```
